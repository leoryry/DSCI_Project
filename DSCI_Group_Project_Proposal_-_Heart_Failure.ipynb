**(2)Clean and wrangle your data into a tidy format**

Since we only consider seven variables, which invovle age, sex, smoking, high_blood_pressure, creatinine_phosphokinase, platelets, and DEATH_EVENT, then we will select these variables.

```r
heart_failure <- select(heart_failure_data, age, sex, smoking, high_blood_pressure, 
                        creatinine_phosphokinase, platelets, DEATH_EVENT)
names(heart_failure) <- c('age','sex','smoking','high_blood_pressure','creatinine_phosphokinase','platelets','DEATH_EVENT') 
heart_failure
```

**(3)Summarize the data in at least one table**

We would summarize creatinine phosphokinase, calculate the average and grouped by sex.

```r
heart_failure_summary <- heart_failure %>% 
   group_by(sex) %>% 
   summarize(average_creatinine_phosphokinase = mean(creatinine_phosphokinase, na.rm = TRUE))
heart_failure_summary
```
